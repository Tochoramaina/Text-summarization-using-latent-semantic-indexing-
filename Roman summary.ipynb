                                         Text Summarization

This involves extracting the gist of a text from a document in order to lessen the length of the said document without changing the core or essence of the meaning of the text. An example is the executive summary of a research paper. We will utilize mathematics and statistical models in summarizing our document.

Text summarrization is grouped into two methodologies:

1. Extaction Based Techniques: It uses mathematical concepts like SVD(Singular Value Decomposition) to extract content such as words, key phrases and sentences that retains thr original meaning of the document thus creating a short executive summary of the text.

2. Abstraction Based Techiniques: This involves complex algorithms learning the language semantics of a text in order to generate a representative summary of the original text without compromising the original meaning of the text.

We will implement our text summarization using extraction based technique where we will leverage on SVD to summarize our document.

               Text Processing

Before we implement document summarrization, we will perform  text wrangling which will involve:

1. Removing special characters and whitespaces in our text.
2. Changing our text from proper case to lower case.
3. Removing empty long spaces between words.
4. Reducing sentences into words and removing words that have insignificant meaning (stopwords) like articles, conjuctions etc and retaining those that are important in our text.
5. Joining the cleaned words into sentences.


```python
stopwords = nltk.corpus.stopwords.words('english')
# a function to clenase our text
def normalize_doc(doc):
    # removing whitespaces and special charactes
    doc = re.sub(r'[^a-zA-Z\s]', "", doc, re.I|re.A)
    # chaing to lower case
    doc = doc.lower()
    doc = doc.strip()
    # tkenizing text
    tokens = nltk.word_tokenize(doc)
    #removing stopwords
    filtered_words = [token for token in tokens if token not in stopwords]
    # joining words
    doc = " ".join(filtered_words)
    return doc
#
normalize_corpus = np.vectorize(normalize_doc)
```

```python
# Appling our function to cleanse our text
sentences = nltk.sent_tokenize(data)
norm_sentence = normalize_corpus(sentences)
```

We will use TFIDVectorizer from sklearn which performs the following functions:


1. Terms Frequency. This involves is the frequency of a word/term in a document.
2. Inverse document frequency. This involves computing by dividing the total number of documents in our corpus by the document frequency for each term and then applying logarithmic scaling to the result.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# initializing TFIDVectorizer
# using normalizing our results using l2 norm
tv = TfidfVectorizer(min_df=0,max_df=1, norm='l2')
dt_matrix = tv.fit_transform(norm_sentence)
# creating a feature matrix which maps a collection of documents into features (row = document, column= feature)
dt_matrix = dt_matrix.toarray()
# getting various vector names in our text
vocabs = tv.get_feature_names_out()
# transposing our matrix
td_matrix = dt_matrix.T
print(td_matrix.shape)
# showing a feature engineered dataframe
pd.DataFrame(np.round(td_matrix, 2), index=vocabs).head(10)
```

          Latent Semantic Indexing

This method utilizes the principle that a hidden structure exists among terms that are related contextually and hence should also be correlated in the same singular space. It utilises SVD(Singular value decomposition) which is a linear algebra technique of factorizing a matrix e.g matrix M. I wont delve into it but, ""Mathematically, the matrix M can be represented using SVD as 
a factorization such that:
Mm*n = Um*m Sm*n V^t n*n


Where we have the following decompositions:
1. U is a m × m unitary matrix such that UTU = Im × m where I is the 
identity matrix. The columns of U indicate left singular vectors.
2. S is a diagonal m × n matrix with positive real numbers on the 
diagonal of the matrix. This is also often also represented as a vector 
of m values which indicate the singular values.
3. VT is a n × n unitary matrix such that VTV = In × n, where I is the identity 
matrix. The rows of V indicate right singular vectors.
This tells us that U and V are orthogonal. The singular values of S are particularly 
important in summarization algorithms. We use SVD particularly for low rank matrix 
approximation, where we approximate the original matrix M with a matrix Mˆ such that 
this new matrix is a truncated version of the original matrix M with a rank k and can 
be represented by SVD as M U ˆ ˆ
SVT = , where ˆ
S is a truncated version of the original S
matrix and now consists of only the top k largest singular values and the other singular 
values are represented by zero""

```python
from scipy.sparse.linalg import svds
# creating a unction that will leverage SVD
def low_rank_svd(matrix, singular_count):
    u, s, vt = svds(matrix, k = singular_count)
    return u,s,vt
```
